### One sample example

```python
trait = "Height"
dmeta_sub = dmeta[dmeta['Trait_name'] == trait].reset_index(drop=True)
dmeta_sub = dmeta_sub.set_index("genotype")
    
# Subset genotypes present in metadata (TRAIN)
# List of 1000 SNPs
df_cols = pd.read_csv("../945_GEA_lfmm/05_outlier_genotypes_1000_" + trait + ".tsv", sep="\t", header=0)
df_cols = df_cols.columns.values

# Train samples for the trait
df_sub = df_train.loc[dmeta_sub.index.values]
print(df_sub.shape)
# 1000 genotypes
df_sub = df_sub[df_cols]
print(df_sub.shape)

# Check if the order of index matches
sum(df_sub.index.values == dmeta_sub.index.values)/len(df_sub.index.values)


# Subset genotypes
upper_limit = 2
iteration = 0
random_state = 250828 + upper_limit + iteration


df_grouped = (
    df_sub
    .groupby(dmeta_sub["POP_SITE"])
    .apply(lambda x: x.sample(min(len(x), upper_limit), random_state=random_state))
)

        
# Calculating frequencies
df_all = df_grouped.groupby('POP_SITE').count()*2
            
df_alleles = df_grouped.groupby('POP_SITE').sum()
            
df_af = df_alleles.div(df_all)
```

```python
# Checking snps with no variation
col_sums = df_af.sum()
cols_to_exclude = col_sums[col_sums==0].index.values.tolist()
df_af = df_af.drop(columns = cols_to_exclude)
print(df_af.shape)
print(len(cols_to_exclude))
```

### Loop

```python
for trait in traits:

    dmeta_sub = dmeta[dmeta['Trait_name'] == trait].reset_index(drop=True)
    dmeta_sub = dmeta_sub.set_index("genotype")
        
    # Subset genotypes present in metadata (TRAIN)
    # List of 1000 SNPs
    df_cols = pd.read_csv("../945_GEA_lfmm/05_outlier_genotypes_1000_" + trait + ".tsv", sep="\t", header=0)
    df_cols = df_cols.columns.values
    
    # Train samples for the trait
    df_sub = df_train.loc[dmeta_sub.index.values]
    print(df_sub.shape)
    # 1000 genotypes
    df_sub = df_sub[df_cols]
    print(df_sub.shape)
    
    # Check if the order of index matches
    sum(df_sub.index.values == dmeta_sub.index.values)/len(df_sub.index.values)

    for upper_limit in [2, 4, 6, 8, 10, 12, 14, 100]:
        #print(upper_limit)
        for iteration in range(10):
        #for iteration in range(2):
            #print(iteration)
        
            # Subset genotypes
            random_state = 250828 + upper_limit + iteration

            df_grouped = (
                df_sub
                .groupby(dmeta_sub["POP_SITE"])
                .apply(lambda x: x.sample(min(len(x), upper_limit), random_state=random_state))
            )

            # Save randomly selected samples 
            df_samples = df_grouped.reset_index().rename(columns = {'level_1' : 'genotype'})[['POP_SITE','genotype']]
            df_samples.to_csv('./genotypes_trees/01_samples_POP_' + trait + '_' + str(upper_limit) + '_' + str(iteration) + '.tsv', sep='\t', index=True, header=True)
            
            # Calculating frequencies
            df_all = df_grouped.groupby('POP_SITE').count()*2
            
            df_alleles = df_grouped.groupby('POP_SITE').sum()
            
            df_af = df_alleles.div(df_all)
            stats_1 = df_af.shape[1]
        
            # Checking snps with no variation
            col_sums = df_af.sum()
            cols_to_exclude = col_sums[col_sums==0].index.values.tolist()
            df_af = df_af.drop(columns = cols_to_exclude)
            stats_2 = df_af.shape[1]

            # Checking snps with nans
            nan_cols = df_af.columns[df_af.isna().any()].tolist()
            df_af = df_af.drop(columns = nan_cols)
            stats_3 = df_af.shape
            print(trait, upper_limit, iteration, stats_1, stats_2, stats_3)
            
            # Saving
            df_af.to_csv('./genotypes/01_frequencies_POP_' + trait + '_' + str(upper_limit) + '_' + str(iteration) + '.tsv', sep='\t', index=True, header=True)
```
