```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# Genetic offset sets

```python
years = ["2031", "2050", "2070", "2090"]
scenarios = ["245"]
clusters = ["West","Central","East"]
```

# Genetic offset PCs

```python
D1 = []

for year in years:
    for scenario in scenarios:

        file_path = "../974_provenances_offset_run/02_run_gradient_forest_" + year + "_" + scenario + ".tsv"
        if os.path.exists(file_path):
            df_ = pd.read_csv(file_path, sep="\t", header=0)
            df_["year"] = year
            df_["scenario"] = scenario
            df_["set"] = "PC"
            df_["cluster"] = "combined"
            D1.append(df_)
dD1 = pd.concat(D1)
print(dD1.shape)
dD1.head()
dD1.columns
```

```python
#dD.to_csv("01_garden_compare.tsv", sep="\t", header=True, index=False)
```

```python
#dD = pd.concat([dD1, dD2], ignore_index=True)
#dD.head()
dD1.to_csv("01_garden_compare_PCs.tsv", sep="\t", header=True, index=False)
```

# Genetic offset 5 vars

### clusters

```python
P2 = []
for year in years:
    for scenario in scenarios:
        for cluster in clusters:
            file_path = "../975_provenances_offset_run_clusters_var3/02_run_gradient_forest_" + cluster + "_" + year + "_" + scenario + ".tsv"
            if os.path.exists(file_path):
                df_ = pd.read_csv(file_path, sep="\t", header=0)
                df_["year"] = year
                df_["scenario"] = scenario
                df_["set"] = "var3"
                df_["cluster"] = cluster
                P2.append(df_)
dP2 = pd.concat(P2)
print(dP2.shape)
dP2.head()
dP2.columns
```

```python
#dP = pd.concat([dP1, dP2], ignore_index=True)
dP2.to_csv("01_garden_compare_var3.tsv", sep="\t", header=True, index=False)
dP2.head()
```
