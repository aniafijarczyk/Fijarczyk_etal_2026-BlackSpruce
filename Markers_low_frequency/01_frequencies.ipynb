```python
dmeta = pd.read_csv("../44_train_and_test/00_combine_samples.tsv", sep="\t", header=0)

# Subset only TRAIN
dmeta = dmeta[dmeta['SET'] == 'TRAIN'].reset_index(drop=True)
#dmeta = dmeta.set_index("genotype")
print(dmeta.shape)
dmeta.head()
```

# Loop for each trait

```python
traits = dmeta['Trait_name'].drop_duplicates().values.tolist()
traits
```

```python
# random 10000
```

```python
np.random.seed(3092025)
df_rand = df.sample(n=100, axis=1)
print(df_rand.shape)
df_rand.head()
```

```python
for trait in traits:

    print(trait)
    # Data subset
    dmeta_sub = dmeta[dmeta['Trait_name'] == trait].reset_index(drop=True)
    dmeta_sub = dmeta_sub.set_index("genotype")
    
    # Subset genotypes present in metadata (TRAIN)
    df_sub = df_rand.loc[dmeta_sub.index.values]
    print(df_sub.shape)

    # Calculating frequencies
    df_all = df_sub.groupby(dmeta_sub['POP_SITE']).count()*2
    
    df_alleles = df_sub.groupby(dmeta_sub['POP_SITE']).sum()
    
    df_af = df_alleles.div(df_all)
    print(df_af.shape)
    stats_1 = df_af.shape[1]

    # Checking snps with no variation
    col_sums = df_af.sum()
    cols_to_exclude = col_sums[col_sums==0].index.values.tolist()
    df_af = df_af.drop(columns = cols_to_exclude)
    print(df_af.shape)
    stats_2 = df_af.shape[1]

    # Checking snps with nans
    nan_cols = df_af.columns[df_af.isna().any()].tolist()
    df_af = df_af.drop(columns = nan_cols)
    stats_3 = df_af.shape
    print(trait, stats_1, stats_2, stats_3)
    

    # Saving
    df_af.to_csv('01_frequencies_POP_100_' + trait + '.tsv', sep='\t', index=True, header=True)

```
