# DATA

```python
cols = ['sample', 'SITE_ID', 'offset_rda', 'marker', 'trait', 
       'POP_SITE', 'POP_ID', 'SET', 'mean', 'sd', 'n', 'group', 'train_group']
```

```python
df1 = pd.read_csv("01B_garden_compare_rda.tsv", sep="\t", header=0)
df1["train_group"] = "combined"
df1 = df1[cols].reset_index(drop=True)
df1["climate"] = "var3"
df1.head()
```

```python
df2 = pd.read_csv("18_garden_clusters.tsv", sep="\t", header=0)
df2 = df2[cols].reset_index(drop=True)
df2["climate"] = "var3"
df2.head()
```

```python
df3 = pd.read_csv("../951_garden_plot/01B_garden_compare_rda.tsv", sep="\t", header=0)
df3["train_group"] = "combined"
df3 = df3[df3["marker"] == "1000"].reset_index(drop=True)
df3 = df3[cols].reset_index(drop=True)
df3["climate"] = "PC3"
df3.head()
```

```python
df4 = pd.read_csv("../996_garden_plot_clusters/18_garden_clusters.tsv", sep="\t", header=0)
df4 = df4[cols].reset_index(drop=True)
df4["climate"] = "PC3"
df4.head()
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
df = df[df["SET"] == "TEST"].reset_index(drop=True)
df["train_clim"] = df.apply(lambda x: x["train_group"] + "_" + x["climate"], axis=1)
print(df.shape)
df.head()
```

# Get paiwise correlations

```python
df_sub = df[(df["SITE_ID"] == "ML") & (df["trait"] == "Height")].reset_index(drop=True)
df_sub.head()
```

```python
factors = df["train_clim"].unique()
pairs = list(itertools.combinations(factors, 2))
fpairs = [ele for ele in pairs if ("var3" in ele[0]) & ("PC3" in ele[1])]
fpairs
```

```python
# Get all unique factors and their pairwise combinations

def get_pearsons_corr(dataset):
    factors = dataset["train_clim"].unique()
    pairs = list(itertools.combinations(factors, 2))
    fpairs = [ele for ele in pairs if ("var3" in ele[0]) & ("PC3" in ele[1])]
    
    results = []
    
    for f1, f2 in fpairs:
        v1 = dataset.loc[dataset["train_clim"] == f1, "offset_rda"].values
        v2 = dataset.loc[dataset["train_clim"] == f2, "offset_rda"].values
    
        # Only compare if both have the same number of observations
        n = min(len(v1), len(v2))
        rho, pval = pearsonr(v1[:n], v2[:n])
        rho, pval = pearsonr(v1, v2)
        results.append({"train_clim1": f1, "train_clim2": f2, "r": rho, "p": pval, "n" : n})
    
    pairwise_corr = pd.DataFrame(results)
    return(pairwise_corr)
```

```python
get_pearsons_corr(df_sub)
```

# Loop

```python
R = []
for site in ["PR","ML","CH","AC"]:
    for trait in df['trait'].drop_duplicates().values:
        df_sub = df[(df["SITE_ID"] == site) & (df["trait"] == trait)].reset_index(drop=True)

        df_corrs = get_pearsons_corr(df_sub)
        df_corrs['SITE_ID'] = site
        df_corrs['trait'] = trait
        R.append(df_corrs)
dR = pd.concat(R, ignore_index=True)
dR.head()
```

### Saving
